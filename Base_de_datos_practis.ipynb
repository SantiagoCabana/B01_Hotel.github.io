```python
from google.colab import drive
import os
from datetime import datetime
import pytz

# Montar Google Drive
drive.mount('/content/drive')

# Definir la zona horaria de Perú
tz = pytz.timezone('America/Lima')

# Obtener la fecha y hora actual en la zona horaria de Perú
now = datetime.now(tz)
timestamp = now.strftime('%Y%m%d_%H%M%S')

# Definir las rutas
backup_folder = '/content/drive/MyDrive/mysql_backups'
backup_file = os.path.join(backup_folder, f'mysql_backup_{timestamp}.back')

# Crear la carpeta para respaldos si no existe
os.makedirs(backup_folder, exist_ok=True)

# Crear un respaldo de la base de datos
!mysqldump -u root --all-databases > /content/backup.sql

# Comprimir el archivo de respaldo y renombrarlo con la fecha y hora
!tar -czvf {backup_file}.tar.gz /content/backup.sql

# Limpiar archivos temporales
!rm /content/backup.sql

# Verificar que el respaldo se ha creado
!ls -l {backup_folder}
```
